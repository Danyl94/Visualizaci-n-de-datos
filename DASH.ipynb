```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
from jupyter_dash import JupyterDash
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
wb_data=px.data.gapminder()
wb_data.head(30)
```

```python
# se crea la aplicaci√≥n
app  =JupyterDash(__name__)
app.title='banco mundial de datos'
```

```python
title=html.H1('Banco Mundial de Datos') # crea el titulo de la pagina 
```

```python
scatter_header=html.H4('GDP and life expentancy evolution by region ')
```

```python
fig=px.scatter(wb_data, x='gdpPercap', y='lifeExp', size='pop', color='continent', animation_frame='year',
              hover_name='country', range_y=[20,90], size_max=60, log_x=True)
```

```python
#@app.callback(Output("scatter-graph", "figure"), [Input("scatter-dropdown", "value")])                                                      

def select_region(region):
    if region:
        subset =wb_data.query(f"continent == '{region}'")
    else:
        subset =wb_data
        
    fig=px.scatter(subset, x='gdpPercap', y='lifeExp', size='pop', color='continent', animation_frame='year',
              hover_name='country', range_y=[20,90], size_max=60, log_x=True)
    return fig
#se crea un grafica 
```

```python


scatter_dropdown =dcc.Dropdown(
   # id='scatter-dropdown',
    options=[
            {'label':'wordwide', 'value': ''},
            {'label':'Asia', 'value': 'Asia'},
            {'label':'Americas', 'value': 'Americas'},
            {'label':'Europe', 'value': 'Europe'},
            {'label':'Africa', 'value': 'Africa'},
            {'label':'Ocean√≠a', 'value': 'Ocean√≠a'}
    ],
    value=''
)
                
```

```python
# meter la grafica creada  a la p√°gina web
scatter_graph =dcc.Graph(figure=fig)
#scatter_graph =dcc.Graph(id='scatter-graph')
```

```python
scatter_row = html.Div(
    children=[
              scatter_header,
              scatter_dropdown,
              scatter_graph      
    ]
)
```

```python
 # cargar el  aplicativo web 
```

```python
map_header =html.H4('evolutcion por pais')
```

```python
@app.callback(Output('map-graph', 'figure'), [Input('map-radio', 'value')])
def select_metric(metric):
    fig= px.choropleth(wb_data, locations='iso_alpha', color=metric,
                   animation_frame='year',  hover_name='country' )
    
```

```python
map_radio =dcc.RadioItems(
    id='map-radio',
    options=[
            {'label':'GDP per capita', 'value': 'gdpPercap'},
            {'label':'Life expectancy', 'value': 'lifeExp'},
            {'label':'Population', 'value': 'pop'}
            
    ],
    value ='gdpPercap'

)
```

```python
map_graph =dcc.Graph(id='map-graph')
```

```python
map_row = html.Div(
    children=[
              map_header,
              map_radio,
              map_graph
        
    ]


)
```

```python
app.layout =html.Div(
    children=[
              title,
              scatter_row,
               map_row
    ]

) # la aplicacion tiene duse√±o 
```

```python
app.run_server(debug=True, mode='inline')
```
